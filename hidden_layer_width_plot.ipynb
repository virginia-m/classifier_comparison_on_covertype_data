```python
import pandas as pd
import numpy as np
from __future__ import division, print_function
import matplotlib.pyplot as plt
%matplotlib inline

file = 'mlp_explore_results/mlp_explore_hidden_layer_sizes_width.csv'
df_totals = pd.read_csv(file, index_col=0)

parameters = df_totals['hidden_layer_sizes'].values
run_times = df_totals['Run Time'].values
run_times = run_times/60.
accuracy = df_totals['Overall Accuracy'].values

fig, ax1 = plt.subplots(figsize=(13, 5))

yerr = np.zeros(np.size(accuracy))
ax1.plot(parameters, accuracy, 'o-', label='Accuracy (%)')
ax1.plot(parameters, run_times+1, 's-', color='Gray', markersize=4,label='Runtime (min)')

ax2 = ax1.twinx()
ax2.plot(parameters, run_times, 's-', color='Gray', markersize=4,label='Runtime (min)')

ax1.tick_params(axis='both', which='major', labelsize=20)
ax2.tick_params(axis='both', which='major', labelsize=20)

ax1.set_ylim([0.55, 0.95])

ax1.set_xlabel('Single Hidden Layer Width', fontsize=20)
ax1.set_ylabel('Accuracy', fontsize=20)
ax2.set_ylabel('Run Time (min)', fontsize=20)

ax1.grid(True, ls='-', alpha=0.2)
ax1.legend(loc='lower right', fontsize=20)
plt.tight_layout()
plt.savefig('hidden_layer_width_v_accuracy.eps')
plt.show()
```
